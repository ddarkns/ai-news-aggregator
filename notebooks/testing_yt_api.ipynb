```python
import requests
import html2text
from typing import Optional

def url_to_markdown(url: str) -> Optional[str]:
    """
    Simple URL to markdown converter using requests and html2text.
    """
    try:
        headers = {
            'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36',
            'Accept': 'text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8',
            'Accept-Language': 'en-US,en;q=0.5',
            'Referer': 'https://www.google.com/',
            'DNT': '1',
            'Connection': 'keep-alive',
            'Upgrade-Insecure-Requests': '1',
        }
        
        response = requests.get(url, headers=headers, timeout=10)
        response.raise_for_status()
        
        # Convert HTML to Markdown
        h = html2text.HTML2Text()
        h.ignore_links = False
        h.ignore_images = False
        markdown = h.handle(response.text)
        
        return markdown
        
    except Exception as e:
        print(f"Error: {e}")
        return None
```

```python
import asyncio
import nest_asyncio
from playwright.async_api import async_playwright

# Apply nest_asyncio to allow nested event loops
# This prevents "RuntimeError: This event loop is already running"
nest_asyncio.apply()

async def _fetch_transcript_async(video_id: str):
    """
    Internal Async function to scrape transcript using Playwright.
    """
    url = f"https://www.youtube.com/watch?v={video_id}"
    full_transcript = []

    async with async_playwright() as p:
        print("🎭 Launching Playwright Browser (Async)...")
        # headless=True is faster; set to False to watch it work
        browser = await p.chromium.launch(headless=True)
        page = await browser.new_page()
        
        try:
            print(f"   Navigating to {url}...")
            await page.goto(url, wait_until="domcontentloaded")
            
            # 1. Click "More" to expand description
            try:
                # First try the newer '...more' button ID
                expand_button = page.locator("#expand")
                if await expand_button.count() > 0:
                    await expand_button.first.click()
                else:
                    # Fallback for different layouts
                    await page.get_by_text("...more").click()
            except Exception:
                print("   Could not find 'More' button, checking if transcript is visible...")

            # 2. Click "Show transcript"
            try:
                # Try specific button accessibility label
                show_transcript = page.locator("button[aria-label='Show transcript']")
                if await show_transcript.count() > 0:
                    await show_transcript.click()
                    print("   Clicked 'Show transcript' (ARIA)")
                else:
                    # Fallback by text
                    await page.get_by_text("Show transcript").click()
                    print("   Clicked 'Show transcript' (Text)")
            except Exception as e:
                print(f"   ⚠️ Failed to click 'Show transcript': {e}")
                return None

            # 3. Wait for transcript container
            transcript_selector = "ytd-transcript-segment-renderer"
            try:
                await page.wait_for_selector(transcript_selector, timeout=5000)
            except:
                print("   ❌ Transcript element never appeared.")
                return None
            
            # 4. Scroll & Scrape (Handle Lazy Loading)
            # We must scroll the container to force YouTube to load all lines
            print("   Scrolling to load full transcript...")
            scroll_container = page.locator("ytd-transcript-renderer #segments-container")
            
            previous_count = 0
            retries = 0
            
            while retries < 5:
                # Count current segments
                segments = await page.locator(transcript_selector).all()
                current_count = len(segments)
                
                if current_count == previous_count and current_count > 0:
                    # No new segments loaded -> We are likely done
                    retries += 1
                else:
                    retries = 0 # Reset retry if we found new lines
                
                previous_count = current_count
                
                # Scroll the last segment into view to trigger loading next batch
                if segments:
                    await segments[-1].scroll_into_view_if_needed()
                
                await asyncio.sleep(0.5) # Short async pause

            # 5. Extract Text
            print(f"   Extracting text from {len(segments)} segments...")
            for segment in segments:
                # .inner_text() is an async method in Playwright async API
                text = await segment.locator(".segment-text").inner_text()
                full_transcript.append(text)

            return " ".join(full_transcript)

        except Exception as e:
            print(f"❌ Playwright Error: {e}")
            return None
        finally:
            await browser.close()

def fetch_transcript_sync(video_id: str):
    """
    Wrapper function to call the async code from your synchronous main.py.
    """
    try:
        # Since nest_asyncio is applied, this runs smoothly even inside other loops
        return asyncio.run(_fetch_transcript_async(video_id))
    except Exception as e:
        print(f"Wrapper Error: {e}")
        return None

if __name__ == "__main__":
    # Test with the German video ID
    vid_id = "xcNwd0B_EL0" 
    result = fetch_transcript_sync(vid_id)
    if result:
        print(f"\n✅ SUCCESS! Transcript Length: {len(result)}")
        print(result[:200])
    else:
        print("\n❌ Failed to get transcript.")
```
